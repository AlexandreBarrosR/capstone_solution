### Distribuição de preço por stream_id

```python
fig_dims = (10, 15)
fig, ax = plt.subplots(figsize=fig_dims,nrows=2)
temp_data = df.groupby("stream_id")["price"].sum().reset_index().sort_values(by="price", ascending=False).head(30)
ax[0].title.set_text("Distribuição dos preços por stream_id")
ax[0].hist(x=temp_data["price"], bins=50)

temp_data = df[df["customer_id"] != 99999].groupby("stream_id")["price"].sum().reset_index().sort_values(by="price", ascending=False).head(30)
ax[1].title.set_text("Distribuição dos preços por stream_id (sem missing values)")
ax[1].hist(x=temp_data["price"], bins=50)
```

```python
streamsA = df.groupby("stream_id")["price"].sum().reset_index().sort_values(by="price", ascending=False).head(10)["stream_id"]
streamsB = df[df["customer_id"] != 99999].groupby("stream_id")["price"].sum().reset_index().sort_values(by="price", ascending=False).head(10)["stream_id"]
fig, ax = plt.subplots(ncols=2, nrows=len(streamsA), figsize=fig_dims)
fig.suptitle("Distribuição de preços dentro dos top 10 stream_id (por preço)")
fig_dims = (20, len(streamsA) * 6)
i=0;
for item in streamsA:
    ax[i,0].hist(x=df[df["stream_id"] == str(item)]["price"], bins=100)
    ax[i,0].title.set_text(item + " (all values)")
    ax[i,1].hist(x=df[df["stream_id"] == str(streamsB[i: i+1].values[0])]["price"], bins=100)
    ax[i,1].title.set_text(streamsB[i: i+1].values[0] + " (without missing values)")
    i = i + 1
```

### number of customers by year-month

```python
x = df[["year","month","customer_id"]]
x = x[~x.duplicated()]
customer_by_yyyymm = x.groupby(["year","month"]).count().reset_index()
del x
```

```python
aux = pd.DataFrame(columns=["month","2017","2018","2019"])
aux["month"] = list(range(1,13))
aux["2017"] = 0
aux["2018"] = 0
aux["2019"] = 0
aux["2017"][customer_by_yyyymm[customer_by_yyyymm["year"] == 2017]["month"] - 1] = customer_by_yyyymm[customer_by_yyyymm["year"] == 2017]["customer_id"]
aux["2018"][customer_by_yyyymm[customer_by_yyyymm["year"] == 2018]["month"] - 1] = customer_by_yyyymm[customer_by_yyyymm["year"] == 2018]["customer_id"]
aux["2019"][customer_by_yyyymm[customer_by_yyyymm["year"] == 2019]["month"] - 1] = customer_by_yyyymm[customer_by_yyyymm["year"] == 2019]["customer_id"]
aux
```

```python
plt.plot(aux["2017"], c='Black', ls='--', marker='s', ms=7, label="2017")
plt.plot(aux["2018"], c='Red', ls='--', marker='s', ms=7, label="2018")
plt.plot(aux["2019"], c='Blue', ls='--', marker='s', ms=7, label="2019")
plt.legend(loc='upper left', bbox_to_anchor=(1,1))
plt.xticks(list(range(0,12)),aux["month"], rotation='vertical')
plt.show()
del aux
```

```python
aux = customer_by_yyyymm
aux["year_month"] = aux["year"] * 100 + aux["month"]
plt.plot(aux[aux["year"] == 2017]["customer_id"], c='Black', ls='--', marker='s', ms=7, label="2017")
plt.plot(aux[aux["year"] == 2018]["customer_id"], c='Red', ls='--', marker='s', ms=7, label="2018")
plt.plot(aux[aux["year"] == 2019]["customer_id"], c='Blue', ls='--', marker='s', ms=7, label="2019")
plt.legend(loc='upper left', bbox_to_anchor=(1,1))
plt.xticks(list(range(0,21)),aux["year_month"], rotation='vertical')
plt.show()
```
